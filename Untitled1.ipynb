```python
from projectwind.ml_data import data_ml
from projectwind.ml_data import add_timestamps
from projectwind.ml_data import pipeline
from projectwind.ml_data import trainer
%load_ext autoreload
%autoreload 2
from prophet import Prophet
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.metrics import mean_absolute_error
```

```python
data =data_ml()
data= add_timestamps(data)
```

```python
def trainer(data):
    pipe =pipeline()
    ref_date_range = pd.date_range(start=start_data, end=end_data,freq='10T')
    for key, df in data.items():
        # Fit the date and set the index
        print(df)
        data[key] = pd.DataFrame(pipe.fit_transform(df), index=ref_date_range)
    return data
```

```python
start_data ="2019-09-01 00:00:00"
end_data = "2021-07-01 00:00:00"
trainer_data = trainer(data)
```
